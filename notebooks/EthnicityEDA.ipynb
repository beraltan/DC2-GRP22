```python
ethnicity_dist2012 = pd.read_excel('qwe/ethnic-groups-by-borough.xlsx', sheet_name = '2012')
ethnicity_dist2013 = pd.read_excel('qwe/ethnic-groups-by-borough.xlsx', sheet_name = '2013')
ethnicity_dist2014 = pd.read_excel('qwe/ethnic-groups-by-borough.xlsx', sheet_name = '2014')
ethnicity_dist2015 = pd.read_excel('qwe/ethnic-groups-by-borough.xlsx', sheet_name = '2015')
ethnicity_dist2016 = pd.read_excel('qwe/ethnic-groups-by-borough.xlsx', sheet_name = '2016')
ethnicity_dist2017 = pd.read_excel('qwe/ethnic-groups-by-borough.xlsx', sheet_name = '2017')
ethnicity_dist2018 = pd.read_excel('qwe/ethnic-groups-by-borough.xlsx', sheet_name = '2018')
ethnicity_dist2019 = pd.read_excel('qwe/ethnic-groups-by-borough.xlsx', sheet_name = '2019')
ethnicity_dist2020 = pd.read_excel('qwe/ethnic-groups-by-borough.xlsx', sheet_name = '2020')

pd.options.mode.chained_assignment = None

allYears = [ethnicity_dist2012, ethnicity_dist2013, ethnicity_dist2014, ethnicity_dist2015, 
 ethnicity_dist2016, ethnicity_dist2017, ethnicity_dist2018, ethnicity_dist2019, ethnicity_dist2020]

for i  in range(len(allYears)):
    yearData = allYears[i]
    del [yearData['Unnamed: 7'],yearData['95% Confidence Interval'], yearData['Unnamed: 9'],
    yearData['Unnamed: 10'], yearData['Unnamed: 11'], yearData['Unnamed: 12']]
    yearData.columns = ['Code', 'Area', 'White', 'Asian', 'Black', 'Mixed/Other', 'Total']
    yearData.drop([0, 1, 2, 35, 45], inplace = True)
    yearData = yearData.iloc[: 32]

    yearData['White'] = pd.to_numeric(yearData['White'], errors='coerce')
    yearData['Asian'] = pd.to_numeric(yearData['Asian'], errors='coerce')
    yearData['Black'] = pd.to_numeric(yearData['Black'], errors='coerce')
    yearData['Mixed/Other'] = pd.to_numeric(yearData['Mixed/Other'], errors='coerce')
    yearData['Total'] = pd.to_numeric(yearData['Total'], errors='coerce')
    
    yearData['W_perc'] = yearData['White'] / yearData['Total']
    yearData['A_perc'] = yearData['Asian'] / yearData['Total']
    yearData['B_perc'] = yearData['Black'] / yearData['Total']
    yearData['M_perc'] = yearData['Mixed/Other'] / yearData['Total']
    allYears[i] = yearData.reset_index(drop = True)
```
