```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = 'data/output_files/new_trust.csv'
trust = pd.read_csv(file)
trust = pd.DataFrame(trust)

trust.head(100)
```

```python
keep_columns = ['Borough', 'Average Score']
trust = trust[keep_columns]
trust.head(100)
```

```python
trust = trust.groupby(['Borough']).mean()
trust.head(100)
```

```python
highest_5 = trust.nlargest(5, 'Average Score')
highest_5
```

```python
lowest_5 = trust.nsmallest(5, 'Average Score')
lowest_5
```

```python
file_1 = 'qwe/Borough_Correlation_and_Trust_Score.csv'
trust_corr = pd.read_csv(file_1)
trust_corr = pd.DataFrame(trust_corr)
trust_corr.head(100)
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(trust_corr['Average Score'], trust_corr['Correlation'], c = 'blue' , label = 'Corr vs. Average Score')

for i, borough in enumerate(trust_corr['Borough']):
    plt.text(trust_corr['Average Score'][i], trust_corr['Correlation'][i], borough)

plt.xlabel('Average Score')
plt.ylabel('Correlation')
plt.title('Scatter Plot of Rate of Compliant UOF Cases vs Trust by Borough')

z = np.polyfit(trust_corr['Average Score'], trust_corr['Correlation'], 1)
p = np.poly1d(z)
plt.plot(trust_corr['Average Score'], p(trust_corr['Average Score']), "r--", label='Best Fit Line')

slope = z[0]
# plt.text(0.623, 0.31, f'Slope: {slope:.2f}', fontsize=12, color='red')

plt.legend()
plt.grid(True)
plt.show()
```
